# Test the Accuray of the baseline model without Pertubation

```python
from preprocessing import *
from baseline import FaceRecognizer
from tqdm import tqdm
```

```python
fr_baseline = FaceRecognizer()
fr_baseline.load_state_dict(torch.load('best_model.pt'))
```

```python

def evaluate(model, data, targets): 
    model.eval()
    confusion_matrix = np.zeros((40,40))
    for i, selected_batch in tqdm(enumerate(test_data)):
        z = model.forward(selected_batch)
#         print(torch.argmax(z))
#         print(targets[i])
        row = targets[i][0].long()
        col = torch.argmax(z)
        confusion_matrix[row.item()][col.item()] += 1
    return confusion_matrix
```

```python
mtx = evaluate(fr_baseline, test_data, test_target)
```

```python
mtx
```

```python
eval_confusion_matrix(mtx)
```
